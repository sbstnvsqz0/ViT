# Desbalance de clases

```python
from utils.ContarClase import contar_clase
```

```python
f = open("Yoga-82/yoga_train.txt");
lines = f.readlines();
f.close();
```

```python
d = {}
for line in lines:
    folder_name,numbers = line.split("/")[0], [int(line.split("/")[1].split(",")[1]),int(line.split("/")[1].split(",")[2]),int(line.split("/")[1].split(",")[3].split("\n")[0])]
    if folder_name not in d.keys():
        d[folder_name] = numbers 
    else:
        if d[folder_name] != numbers:
            print("Error")
            break
```

```python
def suma(d1,d2):
    output = {}
    for key in list(d1.keys()):
        output[key]=d1[key]+d2[key]
    return output
```

```python
class_82 = suma(contar_clase(d,82,"Images"),contar_clase(d,82,"Test"))
class_20 = suma(contar_clase(d,20,"Images"),contar_clase(d,20,"Test"))
class_6 = suma(contar_clase(d,6,"Images"),contar_clase(d,6,"Test"))
```

```python
def dist_clases(x:dict):
    k = list(x.keys())
    v = list(x.values())
    plt.bar(k,v)
    plt.title('Distribución de clases para jerarquía con {} clases'.format(str(len(k))))
    plt.xlabel("Clase")
    plt.ylabel("Cantidad de ejemplos")
    plt.show()
    Min,Max = min(v),max(v)
    index_min,index_max = np.argmin(v),np.argmax(v)
    print("Mínimo (clase,cantidad): ({}); Máximo (clase,cantidad): ({})".format(str(index_min)+","+str(Min),str(index_max)+","+str(Max)))
```

```python
dist_clases(class_82)
```

```python
dist_clases(class_20)
```

```python
dist_clases(class_6)
```

# Evaluación

```python
from predict import predict
from sklearn.metrics import confusion_matrix,accuracy_score
import seaborn as sns
from torchinfo import summary
```

```python
def metrics(real_labels,predicted_labels):
    c_matrix = confusion_matrix(real_labels, predicted_labels,normalize="true")

    sns.heatmap(c_matrix, annot=False,cmap='summer')
    plt.xlabel('Valores predichos')
    plt.ylabel('Valores verdaderos')
    plt.suptitle("Matriz de confusión")
    plt.show()
    print("Accuracy: {}".format(str(accuracy_score(real_labels,predicted_labels))))
```
